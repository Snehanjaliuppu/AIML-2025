```python
import matplotlib.pyplot as plt
import seaborn as sns
```

a. Plot a histogram to visualize the distribution of Salary across employees.
b. Use appropriate bin sizes and axis labels **bold text**.


```python
plt.hist(df['Salary'], bins=10, edgecolor='black')
plt.xlabel('Salary (USD)')
plt.ylabel('Number of Employees')
plt.title('Distribution of Salary')
plt.show()
```

**C.Identify the most frequent salary range**

```python
most_common_salary_range = df['Salary'].value_counts().idxmax()
print(f"The most common salary range is around {most_common_salary_range} USD.")
```

```python
def plot_Histogram(df,column,bins=10):
 plt.figure(figsize=(8,6))
 sns.histplot(df[column],bins=bins,kde=True,color='blue')
 plt.title(f"Histogram of{column}")
 plt.xlabel(column)
 plt.ylabel('Frequency')
 plt.show()
```

```python
def plot_scatter(df,x_column,y_column):
  plt.figure(figsize=(8,6))
  sns.scatterplot(x=df[x_column],y=df[y_column],color='red')
  plt.title(f'Scatter plot of {x_column} vs {y_column}')
  plt.xlabel(x_column)
  plt.ylabel(y_column)
  plt.show()
```

```python
def plot_bar_chart(df, x_column, y_column):
  plt.figure(figsize=(10,6))
  sns.barplot(x=df[x_column],y=df[y_column],color='green')
  plt.title(f'Bar chart of {x_column} vs {y_column}')
  plt.xlabel(x_column)
  plt.ylabel(y_column)
  plt.xticks(rotation=45)
  plt.show()
```

```python
file_path = ('/content/employee_data_250-5.csv')
data = load_data(file_path)
```

```python
plot_Histogram(data, 'Salary')
```

```python
plot_scatter(data,'Age', 'Salary')
```

```python
plot_bar_chart(data,'Department','Salary')
```
